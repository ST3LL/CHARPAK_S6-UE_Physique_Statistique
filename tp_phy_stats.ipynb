# Imports

```
import numpy as np
import matplotlib.pyplot as plt
```

```
def sum_tab(tab):
  t = [tab[0]]
  for i in range(1, len(tab)):
    t.append(t[-1]+ tab[i])
  return np.array(t)
```

#1 Mouvement bronien

## 1.1 Marche al√©atoire 2D

### Pas taille 1

```
n = 1000

angles = 2*np.pi*np.random.rand(n)

pas_x = np.cos(angles)
pas_y = np.sin(angles)
px = sum_tab(pas_x)
py = sum_tab(pas_y)

plt.plot(px, py)
plt.plot(0, 0, "or")
plt.show()
```

### Pas taille random

```
n = 1000

pas = np.random.rand(n)
angles = 2*np.pi*np.random.rand(n)

pas_x = pas * np.cos(angles)
pas_y = pas * np.sin(angles)
px = sum_tab(pas_x)
py = sum_tab(pas_y)

plt.plot(px, py)
plt.plot(0, 0, "or")
plt.show()
```

## 1.2 Marche al√©atoire 2D sur r√©seau

```
n = 100

axs = (2*np.random.randint(0,2,n)-1) * np.sqrt(2)/2
ays = (2*np.random.randint(0,2,n)-1) * np.sqrt(2)/2

px = sum_tab(axs)
py = sum_tab(ays)

plt.plot(px, py)
plt.plot(0, 0, "or")
plt.show()
```

#2 Ensemble de marcheurs 2D

#4 Equation de diffusion

##4.1 Mod√©lisation de la diffusion

*   List item
*   List item



```
nb = 1000

m = np.zeros((100,10000))
#              i    t

m[50,0] = nb
for t in range(m.shape[1]-1):
  for i in range(m.shape[0]):
    m[i,t+1] = m[i,t] - 0.02*m[i,t]
    if (i > 0):
      m[i,t+1] += 0.01*m[i-1,t]
    if (i < m.shape[0]-1):
      m[i,t+1] += 0.01*m[i+1,t]


plt.plot(m[:,0]/max(m[:,0]))
plt.plot(m[:,5000]/max(m[:,5000]))
plt.plot(m[:,9999]/max(m[:,9999]))
plt.show()
```
